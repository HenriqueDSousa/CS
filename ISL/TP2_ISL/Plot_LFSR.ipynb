  Python code to plot the results of a Linear Feedback Shift Register, wich was implemented in Verilog language. It is necessary to, before using, upload the results to the directory.
   
  The first two plotted graphs references the short and the medium length LFSR. The last one references the long LFSR, whose results were plotted through a bitmap.
  

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from PIL import Image as im
```

1) Opening LSFR output and plotting graphs

```python
y_vals1 = np.genfromtxt("output1.out")

x_vals1 = np.arange(0, (np.shape(y_vals1)[0]), 1)

plt.figure(figsize=(15,5))

plt.plot(x_vals1, y_vals1, "-o", color = "red")
plt.suptitle("LSFR 9 bits: $x^5$ + $x^9$ + 1 (250 cycles)")
plt.grid(True)

plt.show()
```

```python
y_vals2 = np.genfromtxt("output2.out")

x_vals2 = np.arange(0, (np.shape(y_vals2)[0]), 1)

plt.figure(figsize=(20,5))

plt.plot(x_vals2, y_vals2, "-o", color = "red")
plt.suptitle("LSFR 11 bits: $x^9$ + $x^11$ + 1 (400 cycles)")
plt.grid(True)

plt.show()
```

2)Opening long LSFR and plotting bitmap

```python
y_vals3 = np.genfromtxt("output3.out")

#771000 cycles
y_vals3 = np.reshape(y_vals3, (600,1285))

plt.imsave('bitmap.jpg', y_vals3, cmap = 'gray')
```
